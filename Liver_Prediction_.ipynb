# Domain Analysis

Total_Bilirubin:Bilirubin is a yellowish substance in your blood. It forms after red blood cells break down, and it travels through your liver, gallbladder, and digestive tract before being excreted.Typically, bilirubin levels fall somewhere between 0.1 and 1.2 milligrams per deciliter (mg/dL). Anything above 1.2 mg/dL is usually considered high.The condition of having high bilirubin levels is called hyperbilirubinemia.Many babies are also born with high bilirubin, causing a condition called newborn jaundice. This causes yellow-tinted skin and eyes. It happens because, at birth, the liver often isn’t yet fully able to process bilirubin. This is a temporary condition that usually resolves on its own within a few weeks.A bilirubin test measures the amount of bilirubin in your blood. It’s used to help find the cause of health conditions like jaundice, anemia, and liver disease.


Direct_Bilirubin:The conjugated (direct) bilirubin level is often elevated by alcohol, infectious hepatitis, drug reactions, and autoimmune disorders. Posthepatic disorders also can cause conjugated hyperbilirubinemia.Bilirubin (bil-ih-ROO-bin) is a yellowish pigment that is made during the normal breakdown of red blood cells. Bilirubin passes through the liver and is eventually excreted out of the body.Higher than normal levels of bilirubin may indicate different types of liver or bile duct problems. Occasionally, higher bilirubin levels may be caused by an increased rate of destruction of red blood cells (hemolysis).Normal results for a total bilirubin test are 1.2 milligrams per deciliter (mg/dL) for adults and usually 1 mg/dL for those under 18. Normal results for direct bilirubin are generally 0.3 mg/dL.Lower than normal bilirubin levels are usually not a concern. Elevated levels may indicate liver damage or disease


Alkaline_Phosphotase:Alkaline phosphatase (ALP) is a protein found in all body tissues. Tissues with higher amounts of ALP include the liver, bile ducts, and bone.Abnormal levels of ALP in your blood most often indicate a problem with your liver, gallbladder, or bones. However, they may also indicate malnutrition, kidney cancer tumors, intestinal issues, a pancreas problem, or a serious infection.the normal range for serum ALP level is 20 to 140 IU/L.The normal range runs higher in children and decreases with age.Higher-than-normal levels of ALP in your blood may indicate a problem with your liver or gallbladder. This could include hepatitis, cirrhosis, liver cancer, gallstones, or a blockage in your bile ducts.

SGPt:SGPT exists predominantly in the liver and leaks into the bloodstream when produced in excess. The SGPT normal range is about 7 to 56 units per liter of blood serum. Thus, very high level of SGPT in the blood can be an indication of damage or problems related to the liver. Certain diseases like cirrhosis and hepatitis raise the blood serum SGPT levels, so do specific medications including statin used to lower cholesterol.

Sgot:The normal range of an SGOT test is generally between 8 and 45 units per liter of serum. In general, men may naturally have higher amounts of AST in the blood. A score above 50 for men and 45 for women is high and may indicate damage


Alamine_Aminotransferase:Alanine aminotransferase (ALT) is an enzyme found mostly in the cells of the liver and kidney. Much smaller amounts of it are also found in the heart and muscles. Normally, ALT levels in blood are low, but when the liver is damaged, ALT is released into the blood and the level increases.The normal value for ALT in blood ranges from 29 to 33 units per liter (IU/L) for males and 19 to 25 IU/L for females.Higher-than-normal levels of ALT can indicate liver damage


Aspartate_Aminotransferase::AST is an enzyme your liver makes. Other organs, like your heart, kidneys, brain, and muscles, also make smaller amounts. AST is also called SGOT .ST levels increase when there’s damage to the tissues and cells where the enzyme is found. AST levels can rise as soon as six hoursTrusted Source after damage to tissue occurs. The normal range for AST is higher from birth to age 3 compared to the normal ranges for older children and adults.The normal range is 8 to 33 U/L.





Total_Protiens The normal range for total protein is between 6 and 8.3 grams per deciliter (g/dL).the total protein test measures the total amount of two classes of proteins found in the fluid portion of your blood. These are albumin and globulin.If your total protein level is low, you may have a liver or kidney problem, or it may be that protein isn't being digested or absorbed properly. 

ALB :An albumin blood test measures the amount of albumin in your blood. Albumin is a protein made by your liver. Albumin helps keep fluid in your bloodstream so it doesn't leak into other tissues. It is also carries various substances throughout your body, including hormones, vitamins, and enzymes. Low albumin levels can indicate a problem with your liver or kidney.The normal range is 3.4 to 5.4 g/dL (34 to 54 g/L).



Albumin_and_Globulin_Ratio:Albumin and globulin are proteins that are naturally found in the serum, the liquid part of your blood that doesn’t include blood cells or clotting components.An imbalance in the ratio of albumin to globulin may signify ongoing inflammation, liver problems, or in rare cases immunodeficiency. There is emerging evidence that a low ratio (less albumin and more globulin) may be associated with the risk of cancer and may also predict worse outcomes in cancer and heart disease patients.In general, an albumin/globulin ratio between 1.1 and 2.5 is considered normal

```python
## Lets check outliers in any
# there are outliers in almost all the featurs.Certain models like Logistic regression models,
#are highly influenced by outliers.So it is better to treat the outlier before using this model,
#or use a different model where outlier treatment is not necessary eg.decision tree.

box=data[['Age', 'TB','DB','Alkphos', 'Sgpt','Sgot','TP','ALB','Albumin_and_Globulin_Ratio']]
plt.figure(figsize=(15,20), facecolor='white')

plotnumber = 1

for column in box:
    if plotnumber<=10 :     
        ax = plt.subplot(9,2,plotnumber)
        sns.boxplot(x=box[column])
        plt.xlabel(column,fontsize=20)
        plt.tight_layout()
    plotnumber+=1
plt.show()
```

```python
data_with_outlier_treatment=data.copy()
```

```python
#Before treating the outlier we need to check is the features are skewed or not because,they
#use different calculation
print("Skewness of TB: %f" % data_with_outlier_treatment['TB'].skew())
print("Skewness of: %f" % data_with_outlier_treatment['DB'].skew())
print("Skewness of Alkphos: %f" % data_with_outlier_treatment['Alkphos'].skew())
print("Skewness of Sgpt: %f" % data_with_outlier_treatment['Sgpt'].skew())
print("Skewness of Sgot: %f" % data_with_outlier_treatment['Sgot'].skew())
print("Skewness of TP : %f" % data_with_outlier_treatment['TP'].skew())
print("Skewness of ALB: %f" % data_with_outlier_treatment['ALB'].skew())
print("Skewness of Albumin_and_Globulin_Ratio: %f" % data_with_outlier_treatment['Albumin_and_Globulin_Ratio'].skew())
#Skewness should be between +- 1.96
```

```python
def for_skewed(feature):
    coloumn_name= feature

    IQR=data[coloumn_name].quantile(0.75)-data[coloumn_name].quantile(0.25)
    lower_bridge=data[coloumn_name].quantile(0.25)-(IQR*3)
    upper_bridge=data[coloumn_name].quantile(0.75)+(IQR*3)
    return upper_bridge,lower_bridge
```

### treating the outlier

```python
#TB
upper_boundary,lower_bridge = for_skewed("TB")
print('TN Upper Boundary:',upper_boundary)
data_with_outlier_treatment.loc[data['TB']>=upper_boundary,'TB']=upper_boundary

#DB
upper_boundary,lower_bridge = for_skewed("DB")
print('DB upper Boundary:',upper_boundary)
data_with_outlier_treatment.loc[data['DB']>=upper_boundary,'DB']=upper_boundary
#there is no outlier for lower boundary so no need of treatment
#Alkphos
upper_boundary,lower_bridge = for_skewed("Alkphos")
print('Alkphos upper Boundary:',upper_boundary)
data.loc[data['Alkphos']>=upper_boundary,'Alkphos']=upper_boundary
#there is no outlier for lower boundary so no need of treatment
#Sgot
upper_boundary,lower_bridge = for_skewed("Sgot")
print('SGOT upper boundary:',upper_boundary)
data_with_outlier_treatment.loc[data['Sgot']>=upper_boundary,'Sgot']=upper_boundary
#there is no outlier for lower boundary so no need of treatment
#Sgpt
upper_boundary,lower_bridge = for_skewed("Sgpt")
print('SGpT upper boundary:',upper_boundary)
data_with_outlier_treatment.loc[data['Sgpt']>=upper_boundary,'Sgpt']=upper_boundary
#there is no outlier for lower boundary so no need of treatment
```

```python
def plot_roc_curve(fpr, tpr):
    plt.plot(fpr, tpr, color='orange', label='ROC')
    plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend()
    plt.show()

fpr, tpr, thresholds = roc_curve(y_test, y_pred)
print("tpr ",tpr)
print("fpr ",fpr)
print("thresholds ",thresholds)
print("roc_auc_score" ,roc_auc_score(y_test, y_pred))
optimal_idx = np.argmax(tpr - fpr)
optimal_threshold = thresholds[optimal_idx]
print("Threshold value is:", optimal_threshold)
plot_roc_curve(fpr, tpr)
```

```python
### model creation
X = data_with_outlier_treatment.drop(columns = ['Output'])
y = data_with_outlier_treatment['Output']
```

```python
## scaling
scalar = StandardScaler()
X_scaled = scalar.fit_transform(X)
```

```python
## training and testing data
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,random_state=42)
```

```python
smote = SMOTE()
X_train_smote, y_train_smote = smote.fit_resample(X_train.astype('float'),y_train)
```

```python
## taking optimal k how many clusters to create
error_rate = []
# Will take some time
for i in range(1,11):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train_smote,y_train_smote)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,11),error_rate,color='blue', linestyle='dashed', 
         marker='o',markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
# let's fit the data into kNN model and see how well it performs:
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train_smote,y_train_smote)
```
